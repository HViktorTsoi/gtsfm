```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
gtsfm_df = pd.read_csv('GTSFM Pipeline Edge List.csv')

gtsfm_graph = nx.from_pandas_edgelist(gtsfm_df, source='From', target='To', edge_attr=True, create_using=nx.DiGraph())
pos = nx.spring_layout(gtsfm_graph, k=1)

nx.draw(gtsfm_graph,pos,edge_color='black',width=1,linewidths=1,\
        node_size=50,node_color='teal',alpha=0.9,\
        labels={node:node for node in gtsfm_graph.nodes()})
```
